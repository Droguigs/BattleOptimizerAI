```python
import pandas as pd
```

```python
text = []
with open('teams/teams.txt', 'r') as f:
    text = f.read()
```

```python
textSections = text.split("\n\n")
textList = []
for section in textSections:
    sectionList = section.split("\n")
    textList.append(sectionList)
```

```python
textList[0]
```

```python
for text in textList:
    print(text)
```

```python
df = pd.DataFrame()
for text in textList:
    section_df = pd.DataFrame()
    if (len(text) > 5):
        level = text[2].split(":")
        level = level[1].replace(" ", "")
        EVs = text[3].split(":")
        EVs = EVs[1].split("/")
        IVs = text[4].split(":")
        IVs = IVs[1].split("/")
        text[5] = text[5].replace("- ", "")
        if (len(text) > 6):
            text[6] = text[6].replace("- ", "")
            text[7] = text[7].replace("- ", "")
            text[8] = text[8].replace("- ", "")
            section_df = pd.DataFrame({
                'Name': [text[0]],
                'Level': [level[1]],
                'EVs': [EVs],
                'IVs': [IVs],
                'Move-1': [text[5]],
                'Move-2': [text[6]],
                'Move-3': [text[7]],
                'Move-4': [text[8]]
            })
        else:
            section_df = pd.DataFrame({
                'Name': [text[0]],
                'Level': [level[1]],
                'EVs': [EVs],
                'IVs': [IVs],
                'Move-1': [text[5]],
                'Move-2': ["NaN"],
                'Move-3': ["NaN"],
                'Move-4': ["NaN"]
            })
        df = pd.concat([df, section_df], ignore_index=True)
```
